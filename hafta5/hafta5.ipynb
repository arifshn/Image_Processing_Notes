```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_original_img = cv2.imread("./siyah1.jpg",0)

my_kernel = np.ones((5,5),dtype=np.uint8)

my_erosion_img = cv2.erode(my_original_img , my_kernel , iterations=3)

fig=plt.figure(figsize=(10,7))
rows=1
columns=2

fig.add_subplot(rows,columns,1)
plt.imshow(my_original_img,cmap="gray")
plt.axis("off")
plt.title("ORIGINAL IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.axis("off")
plt.title("EROSION IMAGE")
```

```python
my_dilation_img = cv2.dilate(my_original_img , my_kernel , iterations=3)

fig=plt.figure(figsize=(10,7))
rows=1
columns=3

fig.add_subplot(rows,columns,1)
plt.imshow(my_original_img,cmap="gray")
plt.axis("off")
plt.title("ORIGINAL IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.axis("off")
plt.title("EROSION IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(my_dilation_img,cmap="gray")
plt.axis("off")
plt.title("DILATION IMAGE")
```

```python
new_img = cv2.imread("./siyah2.jpg",0)
white_nose = np.random.randint(0,2,size=new_img.shape[0:2])
white_nose = white_nose*255
new_img.shape
plt.imshow(white_nose,cmap="gray")
plt.axis("off")
plt.title("WHTE-NOSE IMAGE")
```

```python
my_noised_img = new_img + white_nose
fig=plt.figure(figsize=(10,7))
rows=1
columns=2

fig.add_subplot(rows,columns,1)
plt.imshow(new_img,cmap="gray")
plt.axis("off")
plt.title("ORIGINAL IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(my_noised_img,cmap="gray")
plt.axis("off")
plt.title("NOISED IMAGE")
```

```python
my_opening_img = cv2.morphologyEx(my_noised_img.astype(np.float32),cv2.MORPH_OPEN,my_kernel)
fig=plt.figure(figsize=(10,7))
rows=1
columns=3

fig.add_subplot(rows,columns,1)
plt.imshow(new_img,cmap="gray")
plt.axis("off")
plt.title("ORIGINAL IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(my_noised_img,cmap="gray")
plt.axis("off")
plt.title("NOISED IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(my_opening_img,cmap="gray")
plt.axis("off")
plt.title("OPENING IMAGE")
```

```python
my_black_nose = np.random.randint(0,2,size=my_original_img.shape[:2])
my_black_nose = my_black_nose * -255

plt.imshow(my_black_nose,cmap="gray")
plt.axis("off")
plt.title("BLACK NOISED IMAGE")

```

```python
my_black_noised_img = my_black_nose + new_img
my_black_noised_img[my_black_noised_img <= -245 ] = 0
my_closing_img = cv2.morphologyEx(my_black_noised_img.astype(np.float32),cv2.MORPH_CLOSE,my_kernel)
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=3

fig.add_subplot(rows,columns,1)
plt.imshow(new_img,cmap="gray")
plt.axis("off")
plt.title("ORIGINAL IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(my_black_noised_img,cmap="gray")
plt.axis("off")
plt.title("BLACK NOISED IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(my_closing_img,cmap="gray")
plt.axis("off")
plt.title("CLOSING IMAGE")
```

```python
my_gradient_img_1 = cv2.morphologyEx(my_original_img, cv2.MORPH_GRADIENT,my_kernel)
my_gradient_img_2 = cv2.morphologyEx(new_img, cv2.MORPH_GRADIENT,my_kernel)
```

```python
fig=plt.figure(figsize=(10,7))
rows=2
columns=2

fig.add_subplot(rows,columns,1)
plt.imshow(my_original_img,cmap="gray")
plt.axis("off")
plt.title("ORIGINAL-1 IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(my_gradient_img_1,cmap="gray")
plt.axis("off")
plt.title("GRADIDENT-1 IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(new_img,cmap="gray")
plt.axis("off")
plt.title("ORIGINAL-2 IMAGE")

fig.add_subplot(rows,columns,4)
plt.imshow(my_gradient_img_2,cmap="gray")
plt.axis("off")
plt.title("GRADIENT-2 IMAGE")
```

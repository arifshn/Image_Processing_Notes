```python
width, height,channel = 300, 420, 3
img_flag_germany = np.zeros((width,height,channel),dtype=np.uint8)
img_flag_germany[:100,:]=[0,0,0]
img_flag_germany[100:200,:]=[255,0,0]
img_flag_germany[200:300,:]=[255,255,0]
plt.imshow(img_flag_germany)
plt.axis("off") 
plt.show()
```

```python
width, height,channel = 300, 420, 3
img_flag_russian = np.zeros((width,height,channel),dtype=np.uint8)
img_flag_russian[:100,:]=[255,255,255]
img_flag_russian[100:200,:]=[0,0,255]
img_flag_russian[200:300,:]=[255,0,0]
plt.imshow(img_flag_russian)
plt.axis("off") 
plt.show()
```

```python
fig = plt.figure(figsize=(10,7),facecolor='black')
rows=2
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(img_flag_russian)
plt.axis("off")
plt.title("RUSSIAN FLAG")
```

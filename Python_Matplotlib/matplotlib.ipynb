```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = cv2.imread("./cameraman.jpg",0)

histr = cv2.calcHist([img_gray],[0],None,[256],[0,256])

plt.plot(histr)
plt.show()
```

```python
img_camera = cv2.imread("./cameraman.jpg",0)
img_eq_hist = cv2.equalizeHist(img_camera)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
img_adaptive_eq_hist = clahe.apply(img_camera)

fig = plt.figure(figsize=(10,7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(img_camera, cmap="gray")
plt.axis("off")
plt.title("Original Image")

fig.add_subplot(rows,columns,2)
plt.imshow(img_eq_hist, cmap="gray")
plt.axis("off")
plt.title("HISTOGRAM IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(img_adaptive_eq_hist, cmap="gray")
plt.axis("off")
plt.title("ADAPTVE HISTOGRAM IMAGE")
```

```python
hist_org = cv2.calcHist([img_gray],[0],None,[256],[0,256])
hist_eq = cv2.calcHist([img_gray],[0],None,[256],[0,256])

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(hist_org,color="b")
plt.title("ORIJINAL GORUNTU HISTOGRAMI")

plt.subplot(1,2,2)
plt.plot(hist_eq,color="b")
plt.title("HISTOGRAM ETLEME SONRASI GRNT")
plt.tight_layout()
plt.savefig("histogram_grafikleri.jpg",dpi=300)
plt.show()
```
